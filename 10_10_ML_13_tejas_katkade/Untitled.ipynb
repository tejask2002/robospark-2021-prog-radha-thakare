```python
import pandas as pd       
import numpy as np   
from sklearn import preprocessing 
```

```python

df = pd.read_csv('data.csv')
```

```python
x = df["age"].median()

z = df["body"].mean()

df["age"].fillna(x, inplace = True)
df["cabin"].fillna('unknown', inplace = True)
df["home.dest"].fillna('unknown', inplace = True)
df["boat"].fillna(0, inplace = True)
df["body"].fillna(z, inplace = True)

```

```python

my_label = preprocessing.LabelEncoder()   
df[ 'sex' ]= my_label.fit_transform(df[ 'sex' ])   
df
```

```python
y= df.fare
x=df.drop(['ticket','boat','home.dest','embarked','cabin','name','passenger_id','fare','body','parch','sibsp','survived'],axis=1)
```

```python
x.head(100)
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 0)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier(max_depth=3).fit(X_train,y_train)
print("Training:"+str(clf.score(X_train,y_train)))
print("Test:"+str(clf.score(X_test,y_test)))
pred = clf.predict(X_train)
```
